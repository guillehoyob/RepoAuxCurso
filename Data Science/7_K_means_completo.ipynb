## Introducción

<img width=800 src="https://miro.medium.com/v2/resize:fit:809/0*tamvSiqDneDfw2Vr" width="500">

Los algoritmos de aprendizaje no supervisado son aquellos que buscan patrones y estructuras en los datos sin utilizar etiquetas o información de resultado previamente conocida. Estos algoritmos son útiles cuando no se dispone de datos etiquetados o cuando se quiere explorar la estructura subyacente de los datos.

- Clustering (agrupamiento): Agrupa datos similares en función de sus características.

  - K-means
  - DBSCAN / HDBSCAN
  - Agglomerative Hierarchical Clustering
  - Mean Shift
  - Affinity Propagation
  - Spectral Clustering


- Reducción de dimensionalidad: Reduce la cantidad de variables en un conjunto de datos mientras se conserva la mayor cantidad de información posible.

  - Principal Component Analysis (PCA)
  - t-Distributed Stochastic Neighbor Embedding (t-SNE)
  - Uniform Manifold Approximation and Projection (UMAP)
  - Independent Component Analysis (ICA)
  - Linear Discriminant Analysis (LDA)
  - Autoencoders (para redes neuronales)


- Estimación de densidad: Estima la distribución de probabilidad subyacente de los datos.

  - Kernel Density Estimation (KDE)
  - Gaussian Mixture Models (GMM)


- Detección de anomalías: Identifica datos atípicos o inusuales en un conjunto de datos.

  - Isolation Forest
  - Local Outlier Factor (LOF)
  - One-Class Support Vector Machines (SVM)

El aprendizaje no supervisado es una rama en evolución del aprendizaje automático, y nuevos algoritmos y enfoques continúan siendo desarrollados.

## K-means

<img width=800 src="https://static.javatpoint.com/tutorial/machine-learning/images/k-means-clustering-algorithm-in-machine-learning.png" width="500">

K-means es un método popular de clustering (agrupamiento) en el aprendizaje no supervisado. Su objetivo es agrupar puntos de datos similares en función de sus características y asignarlos a uno de los "K" grupos distintos. El proceso de K-means se basa en la minimización de la suma de las distancias al cuadrado entre los puntos de datos y los centroides de sus respectivos grupos. Los centroides son puntos imaginarios (no necesariamente existentes en el conjunto de datos) que representan el promedio de todos los puntos en un grupo.

Los pasos serían:

- Inicialización: Seleccione "K" puntos de datos aleatoriamente del conjunto de datos como centroides iniciales. El valor de "K" debe ser determinado previamente y puede ser elegido en base a conocimientos del dominio, experimentación o utilizando métodos como el "método del codo".

- Asignación de puntos a los grupos: Para cada punto de datos, calcule la distancia a cada uno de los centroides y asigne el punto al grupo correspondiente al centroide más cercano.

- Actualización de centroides: Calcule el promedio (centroide) de todos los puntos de datos en cada grupo y actualice las posiciones de los centroides.

- Repetir: Repita los pasos 2 y 3 hasta que los centroides no cambien significativamente o se alcance un número máximo de iteraciones predefinido. Esto indica que el algoritmo ha convergido.

El resultado final es un conjunto de "K" grupos, donde cada punto de datos está asignado al grupo de su centroide más cercano. Cabe mencionar que el algoritmo K-means es sensible a la inicialización de los centroides, por lo que puede converger a soluciones subóptimas. Para mitigar esto, se puede ejecutar el algoritmo varias veces con diferentes inicializaciones y seleccionar la solución con el menor error cuadrático total (suma de las distancias al cuadrado entre los puntos de datos y los centroides de sus respectivos grupos).

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# Generar datos sintéticos con 3 grupos
X, y = make_blobs(n_samples = 300, centers=3, random_state=42)

#visualizar
plt.scatter(X[:,0], X[:,1], c=y)
plt.title("Datos generados")
plt.show()
```

```python
y
```

```python
# Crear el algoritmo de KMeans
kmeans = KMeans(n_clusters=3)

#Ajusta el modelo
kmeans.fit(X)

cluster_labels = kmeans.labels_

cluster_labels
```

```python
centroids = kmeans.cluster_centers_
centroids
```

```python
#visualizar
plt.scatter(X[:,0], X[:,1], c=cluster_labels)
plt.scatter(centroids[:,0], centroids[:,1], c='red', marker='x', label="Centroides")
plt.title("Clusters con centroides")
plt.legend()
plt.show()
```

* Qué pasa si cambio el Hyper-parámetro k?

```python
# Definir el rango de valores de k a evaluar
k_values = range(2, 11)

for k in k_values:
    # Crear el algoritmo de KMeans
    kmeans = KMeans(n_clusters=k)

    #Ajusta el modelo
    kmeans.fit(X)

    cluster_labels = kmeans.labels_
    centroids = kmeans.cluster_centers_

    #visualizar
    plt.scatter(X[:,0], X[:,1], c=cluster_labels)
    plt.scatter(centroids[:,0], centroids[:,1], c='red', marker='x', label="Centroides")
    plt.title("Clusters con centroides")
    plt.legend()
    plt.show()
```

* Cuál es el k óptimo?

```python
# Definir el rango de valores de k a evaluar
k_values = range(2, 11)

lista_errores = []
for k in k_values:
    # Crear el algoritmo de KMeans
    kmeans = KMeans(n_clusters=k)

    #Ajusta el modelo
    kmeans.fit(X)

    cluster_labels = kmeans.labels_
    centroids = kmeans.cluster_centers_

    # error cuadrático total
    lista_errores.append(kmeans.inertia_)


sns.lineplot(data=lista_errores).set(title="Buscando a K")
```

```python
optimo_k = k_values[np.argmax(lista_errores)]
print(f"El número óptimo de clusters es: {optimo_k+1}")
```

Coeficiente de silueta: Mide la similitud promedio entre los puntos dentro de un grupo y la similitud promedio con los puntos en otros grupos. Un coeficiente de silueta alto indica una buena separación y cohesión de los grupos.

```python
from sklearn.metrics import silhouette_score
```

```python
# Definir el rango de valores de k a evaluar
k_values = range(2, 11)

lista_errores = []
lista_silueta = []
for k in k_values:
    # Crear el algoritmo de KMeans
    kmeans = KMeans(n_clusters=k)

    #Ajusta el modelo
    kmeans.fit(X)

    cluster_labels = kmeans.labels_
    centroids = kmeans.cluster_centers_

    # error cuadrático total
    lista_errores.append(kmeans.inertia_)
    lista_silueta.append(silhouette_score(X,cluster_labels))


sns.lineplot(data=lista_errores).set(title="Buscando a K")
plt.show()
sns.lineplot(data=lista_silueta).set(title="Buscando a K")
plt.show()
```

## Ventajas

  * K-means es fácil de entender e implementar.
  * K-means es computacionalmente eficiente en comparación con otros algoritmos de clustering, especialmente en conjuntos de datos grandes, ya que tiene una complejidad de tiempo aproximadamente lineal en función del número de puntos de datos
  * K-means siempre converge después de un número finito de iteraciones, aunque puede quedar atrapado en óptimos locales, lo que significa que la solución final puede depender de la inicialización de los centroides.
  * El algoritmo puede adaptarse a conjuntos de datos más grandes utilizando técnicas como el K-means++ para una mejor inicialización de los centroides o el Mini-Batch K-means para procesar subconjuntos de datos en cada iteración.

## Desventajas

* Es necesario especificar el número de grupos (K) antes de ejecutar el algoritmo. Elegir un valor incorrecto de K puede afectar negativamente los resultados del clustering.
* K-means es sensible a la inicialización de los centroides, lo que puede conducir a óptimos locales y resultados inconsistentes en diferentes ejecuciones. Para abordar este problema, se pueden utilizar técnicas como K-means++ o realizar múltiples ejecuciones con inicializaciones aleatorias y seleccionar la solución con la menor distorsión.
* K-means tiende a generar grupos esféricos y de igual tamaño debido a su enfoque basado en la minimización de las distancias al cuadrado. Puede tener dificultades para identificar grupos con formas no esféricas, densidades variables o tamaños diferentes.
* K-means es sensible a valores atípicos, ya que estos pueden afectar significativamente la posición de los centroides y la formación de los grupos. Antes de aplicar K-means, puede ser útil realizar un preprocesamiento de los datos para eliminar o tratar los valores atípicos.
* K-means está diseñado para trabajar con distancias euclidianas, lo que puede no ser apropiado para todos los tipos de datos o problemas. Otros algoritmos de clustering, como K-medoids, permiten el uso de diferentes métricas de distancia.

### Calcular el mejor número de clusters
#### Metodo del Codo

```python
from sklearn.cluster import KMeans

# Lista para almacenar las sumas de los cuadrados de las distancias
sse = []

# Probamos valores de k desde 1 hasta 10
for k in range(1, 11):
    # Inicializamos el modelo KMeans con k clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    # Ajustamos el modelo a los datos
    kmeans.fit(df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])
    # Añadimos la suma de los cuadrados de las distancias al final de la lista
    sse.append(kmeans.inertia_)

# Graficamos la curva del codo
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.title('Método del codo')
plt.xlabel('Número de clusters')
plt.ylabel('Suma de cuadrados de las distancias')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

#### Metodo de la silueta

```python
from sklearn.metrics import silhouette_score

# Lista para almacenar los coeficientes de silueta
silhouette_scores = []

# Probamos valores de k desde 2 hasta 10
for k in range(2, 11):
    # Inicializamos el modelo KMeans con k clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    # Ajustamos el modelo a los datos
    kmeans.fit(df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])
    # Calculamos el coeficiente de silueta
    silhouette_avg = silhouette_score(df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']], kmeans.labels_)
    # Añadimos el coeficiente de silueta a la lista
    silhouette_scores.append(silhouette_avg)

# Graficamos el coeficiente de silueta en función de k
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Método de la silueta')
plt.xlabel('Número de clusters')
plt.ylabel('Coeficiente de silueta promedio')
plt.xticks(range(2, 11))
plt.grid(True)
plt.show()
```

### Ajustar el modelo K-MEANS

```python
# Usaremos el número óptimo de clusters encontrado con el método de la silueta
optimal_k = silhouette_scores.index(max(silhouette_scores)) + 2  # Sumamos 2 porque comenzamos desde k=2

# Inicializamos el modelo KMeans con el número óptimo de clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
# Ajustamos el modelo a los datos
kmeans.fit(df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])
# Obtenemos las etiquetas de los clusters
cluster_labels = kmeans.labels_

# Visualizamos los clusters en un diagrama de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue=cluster_labels, palette='viridis', data=df)
plt.title('Clusters de clientes')
plt.xlabel('Ingreso anual (k$)')
plt.ylabel('Puntuación de gasto (1-100)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```
